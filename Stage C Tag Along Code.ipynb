```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

# Import Webpage

```python
link = urlopen('https://en.wikipedia.org/wiki/The_Croods')
```

```python
#load the website
site = requests.get('https://en.wikipedia.org/wiki/The_Croods')
```

```python
#conversion to beautifulsoup object
bs_object = bs(site.content)
```

```python
#to know the content of the beautifulsoup, the HTML of the page will be printed
content = bs_object.prettify()
print(content)
```

```python
#to get specific info on specific object
bs_object.find(class_ = "infobox vevent")
```

```python
info_box = bs_object.find(class_ = "infobox vevent")
print(info_box.prettify())
```

```python
#extract table rows
info_box.find_all("tr")
site_rows = info_box.find_all("tr")
site_rows
```

```python
#iterate the rows using FOR LOOP
for row in site_rows:
    print (row.prettify())
```

```python
#extract rows to build a dictionary of keys and values
#use enumerate key word to extract the rows
#define a function to find all the list items in those rows

def get_content_value(row):
    if row.find("li"):
        return str([li.get_text() for li in row.find_all("li")])
    else:
        return row.find("td").get_text()
```

```python
film_info = {}
for index, row in enumerate (site_rows):
    if index == 0:
        film_info['Title'] = row.find("th").get_text()
        
    elif index == 1:
        continue
            
    else:
        content_key=row.find("th").get_text()
        content_value=get_content_value(row)
        film_info[content_key]= content_value
                
film_info
```

```python
#clean the extracted data
def get_content_value(row):
    if row.find("li"):
        return str([li.get_text().replace("\xa0", " ").replace("[1]", " ").replace("[2]", " ").replace("[3]", " ").replace("[4]", " ").replace("[5]", " ") for li in row.find_all("li")])
    else:
        return row.find("td").get_text().replace("\xa0", " ").replace("[1]", " ").replace("[2]", " ").replace("[3]", " ").replace("[4]", " ").replace("[5]", " ")
```

```python
cleaned_film_info = {}
for index, row in enumerate (site_rows):
    if index == 0:
        cleaned_film_info['Title'] = row.find("th").get_text(" ", strip=True)
        
    elif index == 1:
        continue
            
    else:
        content_key=row.find("th").get_text(" ", strip=True)
        content_value=get_content_value(row)
        cleaned_film_info[content_key]= content_value
                
cleaned_film_info
```

```python
#converting the dictionary to a dataframe
pd.DataFrame(list(cleaned_film_info.items()))
```
